##### Grouping Data

```python
grouped = data.groupby("Major_category")["Men", "Women"].mean()
grouped 
```

##### Plotting Data

```python
grouped.plot.barh(figsize = (12, 8)) 
plt.savefig("Gender-wise-Dominancy.png")
```

##### Showing Dominancy

```python
def Dominancy(row):
  men = row.Men
  women = row.Women
  if men < women:
    row["Dominancy"] = "Women"
  else:
    row["Dominancy"] = "Men"
  return row 
```

```python
dominant = grouped.apply(Dominancy, axis = 1) 
dominant 
```

```python
dominant = dominant.reset_index()
dominant
```

##### Grouping Men and Women

```python
list_men = []
list_women = []

def check_major(row):
  if row.Dominancy == "Men":
    list_men.append(row.Major_category)
  else:
    list_women.append(row.Major_category)
  return 
```

```python
dominant.apply(check_major, axis = 1)
```

```python
list_men 
```

```python
list_women
```

```python
series_men = pd.Series(list_men, name = "Men Dominancy")
series_men 
```

```python
series_women = pd.Series(list_women, name = "Women Dominancy") 
series_women  
```

#### Task 2 

* Finding top employment in major for men.
* Finding top median(salary) in major for men.


```python
list_men 
```

```python
data
```

```python
men_wise = data.loc[data.Major_category.isin(list_men)]
men_wise  
```

##### Men Major Data With Respect To Employment

```python
emp_men_major = men_wise.sort_values("Employed", ascending = False)
emp_men_major 
```

```python
emp_men_major.groupby("Major_category")["Employed"].max()

# this is number of employees, no need to mean 
```

```python
emp_men_major.groupby("Major_category")["Major", "Employed", "Median"].max()


# This is wrong, because we used max() and it compared Major(which are strings)
# and gave us the wrong result

# We used first because data is sorted w.r.t Employed and we got just the first 
# value of each group(Major_category) we made, it gave us the results because
# we had sorted w.r.t Employed


# If we had not filtered and sorted our data, we could have used max()
```

```python
emp_wise_men_major = emp_men_major.groupby("Major_category")["Major", "Employed", "Median"].first()
emp_wise_men_major 
```

```python
emp_wise_men = emp_wise_men_major.set_index("Major")
emp_wise_men 
```

##### Men Major Data With Respect To Salary 

```python
sal_men_major = men_wise.sort_values("Median", ascending = False)
sal_men_major 
```

```python
sal_wise_men_major = sal_men_major.groupby("Major_category")["Major", "Employed", "Median"].first()
sal_wise_men_major 
```

```python
sal_wise_men = sal_wise_men_major.set_index("Major")
sal_wise_men 
```

##### Normalizing Salary Wise And Plotting It

```python
# The data is irregular: meaning, it is not sorted and there are jumbled/big values.
# We cant make a plot directly for such kind of data.
# We have to normalize this data 
# For that we would take each value and divide by the sum of the column.
# Then multiply it by 100 to get proportion


# Normalized (0-1) # max and min
# Found weightage of each value in the column

sal_wise_men["Employed"][0] / sal_wise_men["Employed"].sum()
```

```python
for i in ["Employed", "Median"]:
  sal_wise_men[i] = (sal_wise_men[i] / sal_wise_men[i].sum()) * 100

sal_wise_men
```

```python
plt.figure(figsize = (8, 6))
sal_wise_men.plot.bar()
plt.savefig("sal-wise-m.png")
```

##### Normalizing Employment Wise And Plotting It

```python
for i in ["Employed", "Median"]:
  emp_wise_men[i] = (emp_wise_men[i] / emp_wise_men[i].sum()) * 100

emp_wise_men
```

```python
plt.figure(figsize = (8, 6))
emp_wise_men.plot.bar()
plt.savefig("emp-wise-m.png")
```
